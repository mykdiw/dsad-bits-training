<font color='gray'> <h3> Description: </h3> </font>

The practical goal of algorithm analysis is to predict the performance of different algorithms in order 
to guide design decisions. 
If I know that the run time of Algorithm A tends to be proportional to the size of the input n, and Algorithm B 
tends to be proportional to n^2, then I expect A to be faster than B for large values of n.
The order of growth is a simple but powerful model of running time. We use just a few structural primitives 
(statements, conditionals, loops, and function calls) to build Python programs, so very often the order of
growth of our programs is one of just a few functions of the problem size. 
These can be classified into: Constant, Logarithmic, Linear, Linearithmic, Quadratic, Cubic, and Exponential.

<font color='gray'> <h3> The objectives of this exercise are: </h3> </font>
- 1. To see how the time taken to execute different types of functions changes with increasing number of elements. 
- 2. To see how the calculated value for each type of equation changes for increasing counts of n

```python
#Constant time complexity: 
#An algorithm is said to have a constant time when it is not dependent 
#on the input data (n). No matter the size of the input data, 
#the running time will always be the same. 

import time
import matplotlib.pyplot as plt
   

#Below is a function that adds two numbers a and b passed to it. 
#The time complexity of this function does not change because it will 
#always add only two digits and it will take equal amount of time to add those digits.

def my_sum(a,b):
    return a+b

#initializing data elements to capture the number of items n and 
#the time taken to execute the function

elements1 = list()
times1 = list()

#run the same function for increating value of integers. 
for i in range(0, 50):
  
    # multiplying by constants to increase the value of the numbers 
    a= 500 * i         
    b= 100000 * i 
    
    start = time.perf_counter()       #start the timer
    my_sum(a, b)               #call the function
    end = time.perf_counter()         #stop the timer after the function has executed
    
    print("Run ", i, " Constant : Add ", a, " and ",b ," in ", end-start)
    
    #record the iteration and the time taken.
    elements1.append(i)
    times1.append(end-start)

# plot the time taken for each of the runs.     
plt.xlabel('n')
plt.ylabel('Time Complexity')
plt.plot(elements1, times1, label ='my_sum')
plt.grid()
plt.legend()
plt.show()
```

```python
#Linear time complexity: 
#An algorithm is said to have a linear time complexity when the running time 
#increases at most linearly with the size of the input data. 
#This is the best possible time complexity when the algorithm must 
#examine all values in the input data. 

import time
import matplotlib.pyplot as plt
  
# Below is a function that squares the value of a particular element in an array 
#The time taken to complete this function will vary linearly depending on the 
#number of elements n, for which the square needs to be calculated. 
#If n is 5 then only 5 values in the array will be squared and the function stops.
#If n is 100, then 100 values in the array will be squared and the function stops.

def my_linear(Arr, n):
    j=0
    while (j < n):
        Arr[j] = Arr[j] * Arr[j];
        j=j+1

#initializing data elements to capture the number of items n and 
#the time taken to execute the function

elements2 = list()
times2 = list()

#the function will be called 50 times and each time the number of elements required to square will be increased by 100. 
for i in range(0, 50):
  

    # initializing an array of 10000 elements. this will remain constant for all iterations. 
    Arr = list(range(i*10000))
    
    
    start = time.perf_counter()      #start the timer
    my_linear(Arr, i*100)      #call the function by adding 100 more elements to n from the previous iteration
    end = time.perf_counter()        #stop the timer after the function has executed
  
    print("Run ", i, "Linear : Array size", len(Arr), " time taken:", end-start)
   
    #record the count of n elements and the corresponding time taken.
    elements2.append(i*100)
    times2.append(end-start)



# plot the time taken for each of the runs.     
plt.xlabel('n')
plt.ylabel('Time Complexity')
plt.plot(elements2, times2, label ='my_linear')
plt.grid()
plt.legend()
plt.show()
```

```python
#Quadratic time complexity: 
#A typical program whose running time has order of growth n^2 has two nested loops, 
#used for some calculation involving all pairs of n elements,
#is said to have quadratic order of growth

import time
import matplotlib.pyplot as plt
  

# Below is a function that has two loops, the outer loop j and inner loop k. 
#The basic operation within the inner loop is to add the value found at 
#position j and position k of the array

def my_quadratic(Arr, n):
    j=0
    while (j < n):
        k=0
        while (k < n):
            Arr[k] = Arr[j] + Arr[k];
            k=k+1
        j=j+1

#initializing data elements to capture the number of items n and 
#the time taken to execute the function

elements3 = list()
times3 = list()

#the function will be called 50 times and each time the number of elements required to square will be increased by 100. 
for i in range(0, 50):
  

   

    # initializing an array of 10000 elements. this will remain constant for all iterations. 
    Arr2 = list(range(i*10000))    
    
    
    start = time.perf_counter()      #start the timer
    my_quadratic(Arr2, i*100)   #call the function by adding 100 more elements to n from the previous iteration
    end = time.perf_counter()         #stop the timer after the function has executed
    
    
    print("Run ", i, "Quadratic: Array size", len(Arr2), " time taken:", end-start)
    
    #record the count of n elements and the corresponding time taken.
    elements3.append(i*100)
    times3.append(end-start)


# plot the time taken for each of the runs.     
plt.xlabel('n')
plt.ylabel('Time Complexity')
plt.plot(elements3, times3, label ='my_quadratic')
plt.grid()
plt.legend()
plt.show() 
```

```python
#Standard Functions
#This program calculates the values of standard functions based on the value of n supplied to it. 
#As the value of n increases, you will see how the calculated value of each of the equations grow differently. 
	
import matplotlib.pyplot as plt


#initializing all lists to store the calculated vlaue of each type of equation.
Value_linear = list()
Value_const_linear = list()
Value_quadratic = list()
Value_cubic = list()
Value_exponential = list()
Value_n = list()

# print the header row 
print('n value, ',' Linear (n), ',' Const Linear (1000*n), ',' Quadratic (n^2), ','Cubic (n^3), ','Exponential (2^n)')

#start a loop to increase the value of n after every iteration
for i in range(1, 10):

    
    #calculate the value of each type of equation based on the value of n (or i in this case)
    n = (i)
    linear = (i) 
    const_linear = (10 * i)
    quadratic = (i * i)
    cubic = (i * i * i)
    exponential = (2 ** i)   # in python the operator to calculate exponential values like 2^n is **
    
    # add each of the freshly calculated values to the list of previously calculated values to plot in the graph
    Value_n.append(n)
    Value_linear.append(linear) 
    Value_const_linear.append(const_linear)
    Value_quadratic.append(quadratic)
    Value_cubic.append(cubic)
    Value_exponential.append(exponential)
    
    #print the newly calculate values to the console 
    print("n value ", n, ":", linear, " , " , const_linear," , " ,  quadratic," , " ,  cubic," , " ,  exponential)
    

#plot the calculated value of each of the equations to represent order of growth.     
plt.xlabel('n value')
plt.ylabel('Order of growth')
plt.plot(Value_n,Value_linear, label ='Linear')
plt.plot(Value_n,Value_const_linear, label ='Const_Linear')
plt.plot(Value_n,Value_quadratic, label ='Quadratic')
plt.plot(Value_n,Value_cubic, label ='Cubic')
plt.plot(Value_n,Value_exponential, label ='Exponential')
plt.grid()
plt.legend()
plt.show()
```
